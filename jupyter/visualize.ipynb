```python
def drawPoints(points_pos, points_color, title, pointSize=0.1):
    plot = k3d.plot(name=title)
    plt_points = k3d.points(positions=points_pos.astype(np.float32), 
                            colors=points_color, 
                            point_size=pointSize, 
                            color_range = [0,1],
                            shader='flat')
    plot += plt_points
    plt_points.shader='3d'
    plot.display()

    
pointSize = 0.5
numPoints = 100000
points = np.random.uniform(0, 256, size=(numPoints, 3))
myColors = np.random.randint(0, 0xFFFFFF, numPoints)

drawPoints(points, myColors, "Cosmo")
```
