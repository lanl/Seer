```python
import ipywidgets as widgets
import random
```

```python
# create fake data
num_metrics = 3
num_ranks = 8
metrics = ["temperature","PAPI_L2_ICM","PAPI_L1_ICM"]

metrics_value = {}
for m in metrics:
    vals = [random.randrange(1, 50, 1) for i in range(8)]
    metrics_value[m]=vals
    
print(metrics_value)
```

```python
w = widgets.Select(
    options=metrics,
    value=metrics[0],
    rows=5,
    description='metric:',
    disabled=False
)

current_value = ""
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        current_value = change['new']
        print(current_value)
        print("changed to %s" % change['new'])

w.observe(on_change)

display(w)

```
